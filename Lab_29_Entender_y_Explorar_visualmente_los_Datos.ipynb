# Seaborn

Seaborn es una biblioteca para hacer gráficos estadísticos en Python. Se basa en matplotlib y se integra estrechamente con las estructuras de datos de pandas.

Seaborn te ayuda a explorar y comprender sus datos. Sus funciones de trazado operan en marcos de datos y matrices que contienen conjuntos de datos completos y realizan internamente el mapeo semántico y la agregación estadística necesarios para producir gráficos informativos. Su API declarativa orientada a conjuntos de datos le permite concentrarse en lo que significan los diferentes elementos de sus gráficos, en lugar de los detalles de cómo dibujarlos.


```python
# Import seaborn
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Apply the default theme
#sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")
tips
```

```python
# Create a visualization

"""
hue:
Variable de agrupación que producirá elementos con diferentes colores. 
Puede ser categórico o numérico, aunque el mapeo de colores se comportará de manera diferente en el último caso.

style:
Variable de agrupación que producirá elementos con diferentes estilos. 
Puede tener un tipo numérico pero siempre será tratado como categórico.
"""

sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

Este gráfico muestra la relación entre cinco variables en el conjunto de datos de `tips` usando una sola llamada a la función `relplot()` de Seaborn. Observe cómo proporcionamos solo los nombres de las variables y sus roles en la gráfica. A diferencia del uso directo de matplotlib, no fue necesario especificar los atributos de los elementos de la trama en términos de valores de color o códigos de marcador. 
Detrás de escena, seaborn manejó la traducción de valores en el marco de datos a argumentos que matplotlib entiende. Este enfoque declarativo le permite concentrarse en las preguntas que desea responder, en lugar de los detalles de cómo controlar matplotlib.

Cuando se estiman los valores estadísticos, Seaborn utilizará bootstrapping para calcular los intervalos de confianza y dibujar barras de error que representan la incertidumbre de la estimación.

La estimación estadística en seaborn va más allá de las estadísticas descriptivas. Por ejemplo, es posible mejorar un diagrama de dispersión al incluir un modelo de regresión lineal (y su incertidumbre) usando `lmplot()`

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```

## Resúmenes de distribución informativos

Los análisis estadísticos requieren conocimiento sobre la distribución de variables en su conjunto de datos. La función seaborn `displot()` admite varios enfoques para visualizar distribuciones. Estos incluyen técnicas clásicas como histogramas y enfoques computacionalmente intensivos como la estimación de la densidad del kernel. En Estadística, la estimación de densidad de Kernel `KDE` es una forma no paramétrica para estimar la función de densidad de probabilidad de una variable aleatoria.

```python
def new_func():
    sns.displot

new_func()(data=tips, x="total_bill", col="time", kde=True)
```

## Plots especializados para datos categóricos

Varios tipos de gráficos especializados en seaborn están orientados hacia la visualización de datos categóricos. Se puede acceder a ellos a través de `catplot()`. Estos plots ofrecen diferentes niveles de granularidad. En el nivel más fino, es posible que desee ver cada observación dibujando un diagrama de "enjambre": un diagrama de dispersión que ajusta las posiciones de los puntos a lo largo del eje categórico para que no se superpongan.

```python
sns.catplot(data=tips, kind="swarm", x="day", y="total_bill", hue="smoker")
```

## Vistas compuestas en conjuntos de datos multivariantes

Algunas funciones combinan múltiples tipos de gráficos para brindar rápidamente resúmenes informativos de un conjunto de datos. Uno, `jointplot()`, se enfoca en una sola relación. Traza la distribución conjunta entre dos variables junto con la distribución marginal de cada variable.

```python
penguins = sns.load_dataset("penguins")
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm")
```

El otro, `pairplot()`, tiene una visión más amplia: muestra distribuciones conjuntas y marginales para todas las relaciones por pares y para cada variable, respectivamente.

```python
sns.pairplot(data=penguins, hue="species")
```

## Relación con matplotlib

Si bien puede ser productivo utilizando solo funciones `Seaborn`, la personalización completa de sus gráficos requerirá cierto conocimiento de los conceptos y la API de `matplotlib`. Por otro lado, los usuarios que vienen de `matplotlib` encontrarán que gran parte de su conocimiento se transfiere.

`Matplotlib` tiene una API completa y poderosa; casi cualquier atributo de la figura se puede cambiar a su gusto. Una combinación de la interfaz de alto nivel de `seaborn` y la capacidad de personalización profunda de `matplotlib` te permitirán explorar rápidamente sus datos y crear gráficos que se pueden adaptar a un producto final con calidad de publicación.

# Entender los datos con Python

```python
# Importamos librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import os
```

```python
# Leemos el archivo
path=os.path.join('Datasets', 'Heart.csv')
df = pd.read_csv(path, delimiter=',')
display(df)
```

## 1.Encuentra las proporciones de población con diferentes tipos de trastornos sanguíneos.

```python
# Contar el valor de cada categoría en una Serie
x = df.Thal.value_counts()
x
```

```python
# Dividimos cada uno de ellos con la población total para encontrar la proporción de población
x / x.sum()
```

```python
# En Thal agregamos Missing y recalculamos la proporción de la población
df["Thal"] = df.Thal.fillna("Missing")
x = df.Thal.value_counts()
x / x.sum()
```

## 2.Encuentre el mínimo, el máximo, la media y la desviación estándar de los datos del colesterol.

```python
# Obtener los descriptivos estadísticos de la variable 'Chol'
df.Chol.describe()
```

## 3.Calcular la proporción poblacional de las personas que tienen el colesterol más alto que la media.

```python
# Obtener la media del colesterol
mean_chol = df.Chol.dropna().mean()
mean_chol
```

```python
len(df[df["Chol"] > mean_chol])
```

```python
# Calcular la proporción poblacional de las personas que tienen el colesterol más alto que la media
len(df[df["Chol"] > mean_chol])/len(df)
```

## 4.Realizar un gráfico de la distribución de los datos del colesterol.

```python
# Crear un histograma para comprender mejor el resultado
sns.histplot(data=df.Chol.dropna())
```

## 5.Representar los datos de colesterol contra el grupo de edad para observar la diferencia en los niveles de colesterol en diferentes grupos de edad de las personas.

```python
# Estadísticos descriptivos
df.Age.describe()
```

```python
# Tabla de frecuencias
my_tab = pd.crosstab(index=df["Age"],  # Creamos la tabla
                    columns="count")  # Nombramos la columna 'count'

my_tab
```

```python
# Creamos una nueva columna en el conjunto de datos que devuelva el número de personas en los diferentes grupos de edad
df["agegrp"]=pd.cut(df.Age, [29,40,50,60,70,80])
df
```

```python
# Crear un boxplot con los grupos de edad en el eje `x` y el nivel de colesterol en el eje `y`
plt.figure(figsize=(12,5))
sns.boxplot(x = "agegrp", y = "Chol", data=df)
```

```python
# El boxplot anterior pero ahora diferenciado según género
df["Sex1"] = df.Sex.replace({1: "Male", 0: "Female"})
plt.figure(figsize=(12, 5))
sns.boxplot(x = "agegrp", y = "Chol", hue = "Sex1", data=df)
```

```python
# Crear un boxplot que represente la población masculina y femenina en función de la edad
sns.boxplot(x = "Sex1", y = "Age", data=df)
```

## 6.Crear una tabla para mostrar el número de personas que tienen cada tipo de dolor torácico ('ChestPain') en cada grupo de edad.

```python
# Tabla de frecuencia según tipo de dolor torácico y rango de edad
df.groupby('agegrp')["ChestPain"].value_counts().unstack()
```

## 7.Realizar la misma tabla que la práctica anterior añadiendo la variable Género.

```python
# Tabla de frecuencia según tipo de dolor torácico por rango de edad y género
dx = df.groupby(["agegrp", "Sex1"])["ChestPain"].value_counts().unstack()
dx
```

## 8.Presentar la proporción de población para cada tipo de dolor torácico en los mismos grupos de la tabla anterior.

```python
# Tabla de frecuencia (%) según tipo de dolor torácico por rango de edad y género
dx1 = dx.apply(lambda x: x/x.sum(), axis=1)
dx1
```

```python
# Igual que la anterior pero da 100% por columna
dx2 = dx.apply(lambda x: x/x.sum(), axis=0)
dx2
```
